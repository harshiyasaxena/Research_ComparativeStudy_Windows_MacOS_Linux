```python
import pandas as pd

df = pd.read_excel("Virtualization_Overhead_dataset.xlsx")

df.head()
```

```python
avg = df.groupby("OS")["Virtualization_Overhead_pct"].mean()

print(avg)
```

```python
virt = [
    avg["Windows"],
    avg["macOS"],
    avg["Linux"]
]
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.bar(["Windows","macOS","Linux"], virt)

plt.title("Virtualization Overhead")

plt.show()
```
