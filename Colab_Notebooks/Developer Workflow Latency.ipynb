```python
import pandas as pd

df = pd.read_excel("Developer_Workflow_dataset.xlsx")

df.head()
```

```python
avg = df.groupby("OS").mean()

avg
```

```python
compile = [
    avg.loc["Windows","Compile_Time_s"],
    avg.loc["macOS","Compile_Time_s"],
    avg.loc["Linux","Compile_Time_s"]
]

docker = [
    avg.loc["Windows","Docker_Time_s"],
    avg.loc["macOS","Docker_Time_s"],
    avg.loc["Linux","Docker_Time_s"]
]

git = [
    avg.loc["Windows","Git_Time_s"],
    avg.loc["macOS","Git_Time_s"],
    avg.loc["Linux","Git_Time_s"]
]
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.scatter(["Windows","macOS","Linux"], compile, label="Compile")
plt.scatter(["Windows","macOS","Linux"], docker, label="Docker")
plt.scatter(["Windows","macOS","Linux"], git, label="Git")

plt.legend()
plt.title("Developer Workflow Latency")

plt.show()
```
