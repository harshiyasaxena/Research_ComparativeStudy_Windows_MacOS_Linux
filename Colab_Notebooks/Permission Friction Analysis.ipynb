```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Permission_Friction_Final.xlsx")

# check data
df.head()
```

```python
win = df[df["OS"]=="Windows"]["Permission_Friction_Score"].values[:5]
mac = df[df["OS"]=="macOS"]["Permission_Friction_Score"].values[:5]
lin = df[df["OS"]=="Linux"]["Permission_Friction_Score"].values[:5]

print(win)
print(mac)
print(lin)
```

```python
plt.figure()

plt.plot(win, label="Windows")
plt.plot(mac, label="macOS")
plt.plot(lin, label="Linux")

plt.legend()
plt.title("Permission Friction Score")

plt.show()
```
