```python
import pandas as pd

df = pd.read_excel("BSII_dataset_realistic.xlsx")

df.head()
```

```python
win = df[df["OS"]=="Windows"]["BSII"].iloc[:5].tolist()
mac = df[df["OS"]=="macOS"]["BSII"].iloc[:5].tolist()
lin = df[df["OS"]=="Linux"]["BSII"].iloc[:5].tolist()
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.plot(win,"o-",label="Windows")
plt.plot(mac,"s-",label="macOS")
plt.plot(lin,"^-",label="Linux")

plt.legend()
plt.title("Background Service Interference Index")

plt.show()
```
