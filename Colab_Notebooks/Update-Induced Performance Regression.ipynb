```python
import pandas as pd

df = pd.read_excel("Update_Regression_dataset.xlsx")

df.head()
```

```python
avg = df.groupby("OS")["Performance_Retention_pct"].mean()

print(avg)
```

```python
after = [
    avg["Windows"],
    avg["macOS"],
    avg["Linux"]
]
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.bar(["Windows","macOS","Linux"], after)

plt.title("Post Update Performance (%)")

plt.show()
```
