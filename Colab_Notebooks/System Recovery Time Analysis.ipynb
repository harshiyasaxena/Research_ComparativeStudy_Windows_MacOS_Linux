```python
import pandas as pd
import matplotlib.pyplot as plt

# Upload your Excel file
from google.colab import files
uploaded = files.upload()

# Read dataset
df = pd.read_excel("System_Recovery_FINAL.xlsx")   # use your file name

# Extract first 5 runs per OS
windows = df[df["OS"]=="Windows"]["Total_Recovery_Time_s"].values[:5]
macos   = df[df["OS"]=="macOS"]["Total_Recovery_Time_s"].values[:5]
linux   = df[df["OS"]=="Linux"]["Total_Recovery_Time_s"].values[:5]

# 🔥 IMPORTANT: round values here (this ensures exact graph)
rw = [round(x) for x in windows]
rm = [round(x) for x in macos]
rl = [round(x) for x in linux]

# Plot (same style as your original)
plt.figure()

plt.plot(rw, label="Windows")
plt.plot(rm, label="macOS")
plt.plot(rl, label="Linux")

plt.legend()
plt.title("System Recovery Time")

plt.show()
```
