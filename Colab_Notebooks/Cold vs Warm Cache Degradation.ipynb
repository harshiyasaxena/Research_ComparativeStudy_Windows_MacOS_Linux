```python
import pandas as pd

df = pd.read_excel("CCD_dataset_realistic.xlsx")

df.head()
```

```python
cold_avg = df.groupby("OS")["Cold_Degradation_pct"].mean()
warm_avg = df.groupby("OS")["Warm_Degradation_pct"].mean()

print(cold_avg)
print(warm_avg)
```

```python
os_order = ["Windows","macOS","Linux"]

cold = [cold_avg[os] for os in os_order]
warm = [warm_avg[os] for os in os_order]
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)

plt.figure()

plt.bar(x - 0.2, cold, 0.4, label="Cold")
plt.bar(x + 0.2, warm, 0.4, label="Warm")

plt.xticks(x, ["Windows","macOS","Linux"])

plt.legend()
plt.title("Cold vs Warm Cache Degradation")

plt.show()
```
